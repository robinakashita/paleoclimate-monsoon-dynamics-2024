```python
%matplotlib inline
import pandas as pd
import xarray as xr
import matplotlib.pyplot as plt
import cartopy.feature as cfeature
from cartopy.mpl.gridliner import LONGITUDE_FORMATTER, LATITUDE_FORMATTER
import cartopy.crs as ccrs
```

```python
filepath_1 = ("/volumes/TOSHIBA/MIT/trace.01-36.22000BP.cam2.PRECT.22000BP_decavgDJF_400BCE.nc")
filepath_2 = ("/volumes/TOSHIBA/MIT/trace.01-36.22000BP.cam2.PRECT.22000BP_decavgMAM_400BCE.nc")
filepath_3 = ("/volumes/TOSHIBA/MIT/trace.01-36.22000BP.cam2.PRECT.22000BP_decavgJJA_400BCE.nc")
filepath_4 = ("/volumes/TOSHIBA/MIT/trace.01-36.22000BP.cam2.PRECT.22000BP_decavgSON_400BCE.nc")

min_lat = -40
max_lat = 40
min_lon = 30
max_lon = 120

DJF_new = xr.open_dataset(filepath_1).sel(time=slice(-17,-15),lon=slice(min_lon, max_lon), lat=slice(min_lat,max_lat)).mean(dim='time')*86400000
MAM_new = xr.open_dataset(filepath_2).sel(time=slice(-2,0),lon=slice(min_lon, max_lon), lat=slice(min_lat,max_lat)).mean(dim='time')*86400000
JJA_new = xr.open_dataset(filepath_3).sel(time=slice(-2,0),lon=slice(min_lon, max_lon), lat=slice(min_lat,max_lat)).mean(dim='time')*86400000
SON_new = xr.open_dataset(filepath_4).sel(time=slice(-2,0),lon=slice(min_lon, max_lon), lat=slice(min_lat,max_lat)).mean(dim='time')*86400000

DJF_old = xr.open_dataset(filepath_1).sel(time=slice(-19,-18),lon=slice(min_lon, max_lon), lat=slice(min_lat,max_lat)).mean(dim='time')*86400000
MAM_old = xr.open_dataset(filepath_2).sel(time=slice(-22,-20),lon=slice(min_lon, max_lon), lat=slice(min_lat,max_lat)).mean(dim='time')*86400000
JJA_old = xr.open_dataset(filepath_3).sel(time=slice(-22,-20),lon=slice(min_lon, max_lon), lat=slice(min_lat,max_lat)).mean(dim='time')*86400000
SON_old = xr.open_dataset(filepath_4).sel(time=slice(-22,-20),lon=slice(min_lon, max_lon), lat=slice(min_lat,max_lat)).mean(dim='time')*86400000

DJF_diff = DJF_new - DJF_old
MAM_diff = MAM_new - MAM_old
JJA_diff = JJA_new - JJA_old
SON_diff = SON_new - SON_old
```

```python
domain = [min_lon, max_lon, min_lat, max_lat] # for just indian (longitude start, longitude end, latitude start, latitude end)

fig, ax = plt.subplots(2,2, figsize=(16,9))

#Row 1, Col.1
ax1=plt.subplot(221, projection=ccrs.PlateCarree())

ax1.add_feature(cfeature.COASTLINE, zorder=2)
ax1.add_feature(cfeature.BORDERS, zorder=2)

# rest of this is for tick formatting
#ax1.set_xticks(np.arange(-180, 180, 10), crs=ccrs.PlateCarree())
#ax1.set_yticks(np.arange(-90, 90, 10), crs=ccrs.PlateCarree())
#lon_formatter = LongitudeFormatter()
#lat_formatter = LatitudeFormatter()
#ax1.xaxis.set_major_formatter(lon_formatter)
#ax1.yaxis.set_major_formatter(lat_formatter)

# making map with data
im = DJF_diff.PRECT.plot(transform=ccrs.PlateCarree(), vmin=-1.5, vmax=1.5, cmap='BrBG')

plt.title('Composite_Difference_DJF', fontsize=12) 
ax1.set_extent(domain, crs=ccrs.PlateCarree())

#Row 1, Col.2
ax1=plt.subplot(222, projection=ccrs.PlateCarree())

ax1.add_feature(cfeature.COASTLINE, zorder=2)
ax1.add_feature(cfeature.BORDERS, zorder=2)

# rest of this is for tick formatting
#ax1.set_xticks(np.arange(-180, 180, 10), crs=ccrs.PlateCarree())
#ax1.set_yticks(np.arange(-90, 90, 10), crs=ccrs.PlateCarree())
#lon_formatter = LongitudeFormatter()
#lat_formatter = LatitudeFormatter()
#ax1.xaxis.set_major_formatter(lon_formatter)
#ax1.yaxis.set_major_formatter(lat_formatter)

# making map with data
im = MAM_diff.PRECT.plot(transform=ccrs.PlateCarree(), vmin=-1.5, vmax=1.5, cmap='BrBG')

plt.title('Composite_Difference_MAM', fontsize=12) 
ax1.set_extent(domain, crs=ccrs.PlateCarree())

#Row 2, Col.1
ax1=plt.subplot(223, projection=ccrs.PlateCarree())

ax1.add_feature(cfeature.COASTLINE, zorder=2)
ax1.add_feature(cfeature.BORDERS, zorder=2)

# rest of this is for tick formatting
#ax1.set_xticks(np.arange(-180, 180, 10), crs=ccrs.PlateCarree())
#ax1.set_yticks(np.arange(-90, 90, 10), crs=ccrs.PlateCarree())
#lon_formatter = LongitudeFormatter()
#lat_formatter = LatitudeFormatter()
#ax1.xaxis.set_major_formatter(lon_formatter)
#ax1.yaxis.set_major_formatter(lat_formatter)

# making map with data
im = JJA_diff.PRECT.plot(transform=ccrs.PlateCarree(), vmin=-1.5, vmax=1.5, cmap='BrBG')

plt.title('Composite_Difference_JJA', fontsize=12) 
ax1.set_extent(domain, crs=ccrs.PlateCarree())


#Row 2, Col.2
ax1=plt.subplot(224, projection=ccrs.PlateCarree())

ax1.add_feature(cfeature.COASTLINE, zorder=2)
ax1.add_feature(cfeature.BORDERS, zorder=2)

# rest of this is for tick formatting
#ax1.set_xticks(np.arange(-180, 180, 10), crs=ccrs.PlateCarree())
#ax1.set_yticks(np.arange(-90, 90, 10), crs=ccrs.PlateCarree())
#lon_formatter = LongitudeFormatter()
#lat_formatter = LatitudeFormatter()
#ax1.xaxis.set_major_formatter(lon_formatter)
#ax1.yaxis.set_major_formatter(lat_formatter)

# making map with data
im = SON_diff.PRECT.plot(transform=ccrs.PlateCarree(), vmin=-1.5, vmax=1.5, cmap='BrBG')

plt.title('Composite_Difference_SON', fontsize=12) 
ax1.set_extent(domain, crs=ccrs.PlateCarree())
```
